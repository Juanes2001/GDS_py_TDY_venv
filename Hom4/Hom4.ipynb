```python
# ═══════════════════════════════════════════════════════════════════════════════
# HW4 — Part 1 : 220 nm SOI Strip Waveguide — FDE Wavelength Sweep
# Course: UNAL-BSU PIC Design — Spring 2025
# Student: Juan E. Rodriguez Ochoa
#
# Copy each cell (delimited by ═══ headers) into its own Jupyter notebook cell.
#
# What this script does:
#   • 2D FDE (Finite Difference Eigenmode) of a 450 nm × 220 nm Si strip waveguide
#     on SiO₂, using Lumerical's dispersive material library (Si, SiO₂ - Palik).
#   • Sweeps wavelength from 1500 nm to 1600 nm (51 points).
#   • Extracts neff (TE00, TM00) at every wavelength.
#   • Post-processes: group index ng, GVD coefficient β₂, birefringence.
#   • Stores all results in HDF5 with full caching (safe to interrupt & resume).
#   • Produces publication-quality plots saved as PDF + PNG.
# ═══════════════════════════════════════════════════════════════════════════════


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 1 — Imports · Lumerical API path · Global parameters                 ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

import sys
import os
import platform
import time
import logging
from pathlib import Path
from datetime import datetime

import numpy as np
import h5py

# ─────────────────────────────────────────────────────────────────────────────
# Lumerical installation path  — SET LUMERICAL_VERSION to your folder name.
# Check: C:\Program Files\Lumerical\  and read the subfolder (e.g. "v202").
# ─────────────────────────────────────────────────────────────────────────────
LUMERICAL_VERSION = "v202"          # ← confirmed from your traceback

if platform.system() == "Windows":
    LUMERICAL_ROOT = rf"C:\Program Files\Lumerical\{LUMERICAL_VERSION}"
    LUMERICAL_API  = rf"{LUMERICAL_ROOT}\api\python"
    LUMERICAL_BIN  = rf"{LUMERICAL_ROOT}\bin"
else:
    LUMERICAL_ROOT = f"/opt/lumerical/{LUMERICAL_VERSION}"
    LUMERICAL_API  = f"{LUMERICAL_ROOT}/api/python"
    LUMERICAL_BIN  = f"{LUMERICAL_ROOT}/bin"

# Remove stale cached import if present
if "lumapi" in sys.modules:
    del sys.modules["lumapi"]

if LUMERICAL_API not in sys.path:
    sys.path.insert(0, LUMERICAL_API)

# Required on Python 3.8+ (Windows): os.add_dll_directory for DLL resolution
if platform.system() == "Windows" and hasattr(os, "add_dll_directory"):
    os.add_dll_directory(str(LUMERICAL_BIN))

assert Path(LUMERICAL_API).exists(), (
    f"Lumerical API not found:\n  {LUMERICAL_API}\n"
    f"  Check LUMERICAL_VERSION = '{LUMERICAL_VERSION}'"
)
assert Path(LUMERICAL_BIN).exists(), f"Lumerical bin not found:\n  {LUMERICAL_BIN}"

import lumapi  # noqa — must come after all path setup
print(f"lumapi imported successfully from:\n  {lumapi.__file__}")

# ─────────────────────────────────────────────────────────────────────────────
# Logging
# ─────────────────────────────────────────────────────────────────────────────
logging.basicConfig(
    level   = logging.INFO,
    format  = "%(asctime)s │ %(levelname)s │ %(message)s",
    datefmt = "%H:%M:%S",
)
log = logging.getLogger("HW4_SOI_FDE")

# ─────────────────────────────────────────────────────────────────────────────
# I/O paths
# ─────────────────────────────────────────────────────────────────────────────
VERSION_NAME = "HW4_SOI_220nm_strip_FDE_V1"
DATA_DIR     = Path.cwd() / "data_HW4_SOI_mode_analysis"
DATA_DIR.mkdir(parents=True, exist_ok=True)
HDF5_PATH    = DATA_DIR / f"{VERSION_NAME}.h5"

# ─────────────────────────────────────────────────────────────────────────────
# 220 nm SOI platform — waveguide cross-section
#
#   Y-Z plane (X = propagation direction):
#
#     ┌──────────────────────────────────────────┐
#     │       SiO₂ top cladding (2 µm)           │   z > +WG_H/2
#     │    ┌───────────────────┐                  │
#     │    │       Si core     │  WG_H = 220 nm   │   z ∈ [-110, +110] nm
#     │    └───────────────────┘                  │
#     │       SiO₂ BOX (3 µm)                     │   z < -WG_H/2
#     └──────────────────────────────────────────┘
#              ←── WG_W = 450 nm ──→
#
# A 450 nm wide, 220 nm tall Si strip waveguide on SiO₂ supports a single
# TE-like mode (TE00) and a single TM-like mode (TM00) at 1550 nm.
# ─────────────────────────────────────────────────────────────────────────────
WG_WIDTH_NM  = 450.0     # strip width                 [nm]
WG_HEIGHT_NM = 220.0     # SOI device layer thickness  [nm]

# Simulation domain margins
SIM_Z_BELOW_UM  = 3.0    # SiO₂ BOX depth below core  [µm]
SIM_Z_ABOVE_UM  = 2.0    # SiO₂ top cladding above     [µm]
SIM_Y_MARGIN_UM = 3.0    # lateral cladding each side   [µm]

# FDE mode settings
N_MODES_REQUEST = 4      # request 4 modes — captures TE00, TM00 and 1 order each

# ─────────────────────────────────────────────────────────────────────────────
# Material names from Lumerical's built-in dispersive material database.
# These automatically provide n(λ) at each simulation wavelength.
#
# To verify: open Lumerical MODE → Edit → Materials → search for these names.
# Alternative names that may appear in older versions:
#   "Si (Silicon) - CRC"   /   "Si3N4 (Silicon Nitride)"   /   "SiO2"
# ─────────────────────────────────────────────────────────────────────────────
SI_MATERIAL   = "Si (Silicon) - Palik"     # n ≈ 3.478 at 1550 nm (bulk)
SIO2_MATERIAL = "SiO2 (Glass) - Palik"    # n ≈ 1.444 at 1550 nm

# ─────────────────────────────────────────────────────────────────────────────
# Wavelength sweep: 1500 – 1600 nm, 51 uniformly-spaced points (2 nm step).
# 51 points gives high-quality numerical derivatives (ng, β₂).
# ─────────────────────────────────────────────────────────────────────────────
LAM_START_NM  = 1500.0
LAM_STOP_NM   = 1600.0
N_WAVELENGTHS = 51
SWEEP_WL_NM   = np.linspace(LAM_START_NM, LAM_STOP_NM, N_WAVELENGTHS)  # [nm]
SWEEP_WL_M    = SWEEP_WL_NM * 1e-9                                       # [m] SI
SWEEP_WL_UM   = SWEEP_WL_NM * 1e-3                                       # [µm]

# Physical constant
C_LIGHT_MS = 2.99792458e8   # speed of light [m/s]

# ─────────────────────────────────────────────────────────────────────────────
# Derived geometry [µm]
# ─────────────────────────────────────────────────────────────────────────────
WG_W_UM      = WG_WIDTH_NM  * 1e-3
WG_H_UM      = WG_HEIGHT_NM * 1e-3
_half_h_um   = WG_H_UM / 2.0

SIM_Y_SPAN_UM  = WG_W_UM + 2.0 * SIM_Y_MARGIN_UM             # 6.45 µm
SIM_Z_SPAN_UM  = SIM_Z_BELOW_UM + WG_H_UM + SIM_Z_ABOVE_UM   # 5.22 µm
SIO2_Z_CTR_UM  = -(_half_h_um + SIM_Z_BELOW_UM / 2.0)         # −1.61 µm
SIM_Z_CTR_UM   = (SIM_Z_ABOVE_UM - SIM_Z_BELOW_UM) / 2.0      # −0.50 µm

# ─────────────────────────────────────────────────────────────────────────────
# Mesh sizing: ≥ 15 cells per wavelength inside Si (n ≈ 3.48).
# Step ≈ λ_min / (n_Si × 15) ≈ 1500/(3.48×15) ≈ 28.7 nm — resolves sub-nm
# mode shape variation.  Raise MESH_CELLS_PER_WVL to 20 for publication grade.
# ─────────────────────────────────────────────────────────────────────────────
MESH_CELLS_PER_WVL  = 15
_N_SI_APPROX        = 3.48      # used ONLY for mesh sizing (not simulation)
_mesh_step_um       = SWEEP_WL_UM.min() / (_N_SI_APPROX * MESH_CELLS_PER_WVL)
MESH_CELLS_Y        = int(np.ceil(SIM_Y_SPAN_UM  / _mesh_step_um))
MESH_CELLS_Z        = int(np.ceil(SIM_Z_SPAN_UM  / _mesh_step_um))

# ─────────────────────────────────────────────────────────────────────────────
# Guided-mode cutoff: Re(neff) > N_SIO2_CUTOFF (≈ n_SiO₂ at 1550 nm).
# Modes with Re(neff) below this are radiation/substrate modes — discarded.
# ─────────────────────────────────────────────────────────────────────────────
N_SIO2_CUTOFF = 1.44    # SiO₂ index ≈ 1.444 at 1550 nm; 1.44 is a safe floor.

# ─────────────────────────────────────────────────────────────────────────────
# Print configuration summary
# ─────────────────────────────────────────────────────────────────────────────
print("=" * 62)
print("  220 nm SOI Strip Waveguide — FDE Wavelength Sweep")
print("=" * 62)
print(f"  Core     : {SI_MATERIAL}")
print(f"             {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  (w × h)")
print(f"  Cladding : {SIO2_MATERIAL}  (top + BOX)")
print(f"  λ sweep  : {LAM_START_NM:.0f} → {LAM_STOP_NM:.0f} nm  "
      f"({N_WAVELENGTHS} pts, Δλ = {(LAM_STOP_NM-LAM_START_NM)/(N_WAVELENGTHS-1):.1f} nm)")
print(f"  Domain Y : {SIM_Y_SPAN_UM:.2f} µm  ({MESH_CELLS_Y} cells)")
print(f"  Domain Z : {SIM_Z_SPAN_UM:.2f} µm  ({MESH_CELLS_Z} cells)")
print(f"  Mesh step: ≈ {_mesh_step_um*1e3:.1f} nm  "
      f"(≥ {MESH_CELLS_PER_WVL} cells/λ in Si)")
print(f"  HDF5     : {HDF5_PATH}")
print("=" * 62)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 2 — FDE geometry builder · results extractor · HDF5 init · main sweep ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

# ─────────────────────────────────────────────────────────────────────────────
# Helper 1: _build_fde_SOI
#
#   Creates the FDE solver region and three material regions in a clean session.
#   Uses named Lumerical materials (SI_MATERIAL, SIO2_MATERIAL) so that the
#   solver automatically fetches the correct dispersive n(λ) at each wavelength.
#
#   Structure priority order (higher overrides lower in overlapping regions):
#     1. SiO₂_bg       — fills entire domain (background, lowest priority)
#     2. Si_core        — waveguide rectangle (highest priority)
#   Note: BOX and top cladding are both SiO₂, so SiO₂_bg alone is sufficient;
#   no separate BOX rectangle needed.
# ─────────────────────────────────────────────────────────────────────────────
def _build_fde_SOI(
    mode,
    wavelength_m    : float,
    wg_w_m          : float = WG_WIDTH_NM  * 1e-9,
    wg_h_m          : float = WG_HEIGHT_NM * 1e-9,
    sim_y_span_um   : float = SIM_Y_SPAN_UM,
    sim_z_span_um   : float = SIM_Z_SPAN_UM,
    sim_z_ctr_um    : float = SIM_Z_CTR_UM,
    mesh_cells_y    : int   = MESH_CELLS_Y,
    mesh_cells_z    : int   = MESH_CELLS_Z,
    n_modes_request : int   = N_MODES_REQUEST,
) -> None:
    """
    Build the 220 nm SOI FDE geometry in a clean Lumerical MODE session.
    Uses dispersive material library entries — n evaluated at wavelength_m.

    Parameters
    ----------
    mode           : open lumapi.MODE session
    wavelength_m   : simulation wavelength [m]
    """
    m = mode
    m.switchtolayout()
    m.selectall()
    m.delete()

    # ── FDE solver region ────────────────────────────────────────────────────
    m.addfde()
    m.set("solver type",           "2D X normal")
    m.set("x",                     0.0)
    m.set("y",                     0.0)
    m.set("z",                     sim_z_ctr_um * 1e-6)
    m.set("y span",                sim_y_span_um * 1e-6)
    m.set("z span",                sim_z_span_um * 1e-6)
    m.set("wavelength",            wavelength_m)
    m.set("number of trial modes", n_modes_request)
    m.set("mesh cells y",          mesh_cells_y)
    m.set("mesh cells z",          mesh_cells_z)

    # ── SiO₂ background — fills domain (top cladding, BOX, lateral cladding) ─
    m.addrect()
    m.set("name",    "SiO2_bg")
    m.set("x",       0.0); m.set("x span", 1.0e-6)
    m.set("y",       0.0); m.set("y span", sim_y_span_um * 1e-6)
    m.set("z",       sim_z_ctr_um * 1e-6)
    m.set("z span",  sim_z_span_um * 1e-6)
    m.set("material", SIO2_MATERIAL)   # dispersive SiO₂ from Lumerical library

    # ── Si waveguide core — overrides background in core region ──────────────
    m.addrect()
    m.set("name",    "Si_core")
    m.set("x",       0.0); m.set("x span", 1.0e-6)
    m.set("y",       0.0); m.set("y span", wg_w_m)
    m.set("z",       0.0); m.set("z span", wg_h_m)
    m.set("material", SI_MATERIAL)     # dispersive Si from Lumerical library


# ─────────────────────────────────────────────────────────────────────────────
# Helper 2: _extract_modes_SOI
#
#   After findmodes(), reads neff (complex), TE polarization fraction.
#   Returns arrays of length n_modes_request — NaN where mode was not found.
#
#   Guided criterion: Re(neff) > N_SIO2_CUTOFF
# ─────────────────────────────────────────────────────────────────────────────
def _extract_modes_SOI(mode, n_modes_request: int = N_MODES_REQUEST):
    """
    Extract neff and TE fraction for up to n_modes_request modes from the FDE.

    Returns
    -------
    neff_real : np.ndarray (n_modes_request,)  Re(neff)        NaN if absent
    neff_imag : np.ndarray (n_modes_request,)  Im(neff)
    te_frac   : np.ndarray (n_modes_request,)  TE polarisation fraction [0,1]
    n_guided  : int                            count of guided modes found
    """
    neff_real = np.full(n_modes_request, np.nan, dtype=np.float64)
    neff_imag = np.full(n_modes_request, np.nan, dtype=np.float64)
    te_frac   = np.full(n_modes_request, np.nan, dtype=np.float64)
    n_guided  = 0

    for m_i in range(1, n_modes_request + 1):
        grp = f"FDE::data::mode{m_i}"
        try:
            raw_neff = mode.getdata(grp, "neff")
            raw_te   = mode.getdata(grp, "TE polarization fraction")
        except Exception:
            break   # no more modes available

        neff_c = complex(np.asarray(raw_neff).flat[0])
        te_v   = float(np.asarray(raw_te).flat[0])

        neff_real[m_i - 1] = neff_c.real
        neff_imag[m_i - 1] = neff_c.imag
        te_frac  [m_i - 1] = te_v

        if neff_c.real > N_SIO2_CUTOFF:
            n_guided += 1

    return neff_real, neff_imag, te_frac, n_guided


# ─────────────────────────────────────────────────────────────────────────────
# Helper 3: _init_hdf5_SOI
#   Creates a fresh HDF5 file with pre-allocated NaN datasets and metadata.
#   Layout:
#     /metadata/  — scalar attrs + coordinate datasets
#     /results/   — neff_real, neff_imag, te_frac, n_guided  [N_wl × N_modes]
#     /flags/     — computed  [N_wl] bool
# ─────────────────────────────────────────────────────────────────────────────
def _init_hdf5_SOI(path: Path, n_wl: int, n_modes: int,
                   sweep_wl_nm: np.ndarray, metadata: dict) -> None:
    with h5py.File(path, "w") as f:
        # metadata group
        md = f.create_group("metadata")
        md.create_dataset("sweep_wavelengths_nm", data=sweep_wl_nm)
        md.create_dataset("sweep_wavelengths_m",  data=sweep_wl_nm * 1e-9)
        for k, v in metadata.items():
            md.attrs[k] = v
        md.attrs["timestamp_start"] = datetime.now().isoformat()

        # pre-allocated NaN result datasets — shape [n_wl, n_modes]
        nan2d  = np.full((n_wl, n_modes), np.nan, dtype=np.float64)
        chunk  = (1, n_modes)
        rg = f.create_group("results")
        rg.create_dataset("neff_real", data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("neff_imag", data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("te_frac",   data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("n_guided",
                          data=np.zeros(n_wl, dtype=np.int32),
                          chunks=(1,))

        # progress flag
        f.create_group("flags").create_dataset(
            "computed", data=np.zeros(n_wl, dtype=bool), chunks=(1,)
        )


# ─────────────────────────────────────────────────────────────────────────────
# Main function: run_fde_wavelength_sweep
# ─────────────────────────────────────────────────────────────────────────────
def run_fde_wavelength_sweep(
    sweep_wl_nm     = SWEEP_WL_NM,
    hdf5_path       = HDF5_PATH,
    version_name    = VERSION_NAME,
    hide_gui        = False,    # True for headless / HPC
) -> dict:
    """
    2D FDE wavelength sweep for a 220 nm SOI strip waveguide.

    Uses Lumerical's dispersive material models (SI_MATERIAL, SIO2_MATERIAL)
    so each wavelength step uses the correct n(λ) — no fixed-index approximation.

    Sweep axis: wavelength (1500 → 1600 nm, N_WAVELENGTHS points).
    At each λ:  1. Updates FDE wavelength.
                2. Runs mesh → findmodes().
                3. Extracts neff (real + imag) and TE fraction for all modes.
                4. Writes results to HDF5, flushes (fault-safe).

    Re-running skips already-computed wavelength points (HDF5 cache check).

    Returns
    -------
    dict with keys:
        neff_real    np.ndarray [N_wl, N_modes]
        neff_imag    np.ndarray [N_wl, N_modes]
        te_frac      np.ndarray [N_wl, N_modes]
        n_guided     np.ndarray [N_wl]
        computed     np.ndarray [N_wl]  bool
        sweep_wl_nm  np.ndarray [N_wl]  nm
    """
    n_wl  = len(sweep_wl_nm)
    n_mod = N_MODES_REQUEST
    sweep_wl_m = sweep_wl_nm * 1e-9

    # ── Allocate in-memory result arrays ──────────────────────────────────────
    neff_real = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    neff_imag = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    te_frac   = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    n_guided  = np.zeros(n_wl, dtype=np.int32)
    computed  = np.zeros(n_wl, dtype=bool)

    # ── HDF5 cache load or init ───────────────────────────────────────────────
    if hdf5_path.exists():
        log.info(f"Cache found → {hdf5_path}")
        with h5py.File(hdf5_path, "r") as f:
            if "results/neff_real" in f:
                neff_real[:] = f["results/neff_real"][:]
                neff_imag[:] = f["results/neff_imag"][:]
                te_frac[:]   = f["results/te_frac"][:]
                n_guided[:]  = f["results/n_guided"][:]
                computed[:]  = f["flags/computed"][:]
        n_cached  = int(computed.sum())
        remaining = n_wl - n_cached
        log.info(f"Cached: {n_cached}/{n_wl}  |  Remaining: {remaining}")
        if remaining == 0:
            log.info("All points already computed — returning cached results.")
            return dict(neff_real=neff_real, neff_imag=neff_imag,
                        te_frac=te_frac, n_guided=n_guided,
                        computed=computed, sweep_wl_nm=sweep_wl_nm)
    else:
        log.info("No cache found — initialising new HDF5 file.")
        _init_hdf5_SOI(
            path       = hdf5_path,
            n_wl       = n_wl,
            n_modes    = n_mod,
            sweep_wl_nm= sweep_wl_nm,
            metadata   = {
                "version_name"   : version_name,
                "wg_width_nm"    : WG_WIDTH_NM,
                "wg_height_nm"   : WG_HEIGHT_NM,
                "si_material"    : SI_MATERIAL,
                "sio2_material"  : SIO2_MATERIAL,
                "sim_z_below_um" : SIM_Z_BELOW_UM,
                "sim_z_above_um" : SIM_Z_ABOVE_UM,
                "sim_y_margin_um": SIM_Y_MARGIN_UM,
                "mesh_cells_y"   : MESH_CELLS_Y,
                "mesh_cells_z"   : MESH_CELLS_Z,
                "n_modes_request": n_mod,
                "n_sio2_cutoff"  : N_SIO2_CUTOFF,
            },
        )

    # ── Open Lumerical MODE session ───────────────────────────────────────────
    log.info("Launching Lumerical MODE …")
    mode     = lumapi.MODE(hide=hide_gui)
    t_start  = time.time()
    runs_done  = 0
    runs_total = int((~computed).sum())

    try:
        # Build geometry once at the first uncached wavelength;
        # only the FDE wavelength property is updated in the loop.
        first_uncached = int(np.where(~computed)[0][0])
        _build_fde_SOI(mode, wavelength_m=float(sweep_wl_m[first_uncached]))
        log.info(f"Geometry initialised  ({runs_total} wavelength points remaining) …")

        with h5py.File(hdf5_path, "r+") as hf:
            for wl_idx, wl_m in enumerate(sweep_wl_m):
                if computed[wl_idx]:
                    continue

                # Update wavelength (geometry unchanged)
                mode.switchtolayout()
                mode.setnamed("FDE", "wavelength", float(wl_m))

                # Run FDE mesh + find modes
                try:
                    mode.run()
                    mode.findmodes()
                except Exception as exc:
                    log.warning(
                        f"  FDE FAILED │ λ = {sweep_wl_nm[wl_idx]:.2f} nm │ {exc}"
                    )
                    computed[wl_idx]             = True
                    hf["flags/computed"][wl_idx] = True
                    hf.flush()
                    continue

                # Extract modes
                nr, ni, tf, ng = _extract_modes_SOI(mode, n_mod)

                # Store in memory
                neff_real[wl_idx, :] = nr
                neff_imag[wl_idx, :] = ni
                te_frac  [wl_idx, :] = tf
                n_guided [wl_idx]    = ng
                computed [wl_idx]    = True

                # Write to HDF5 immediately (fault-safe)
                hf["results/neff_real"][wl_idx, :] = nr
                hf["results/neff_imag"][wl_idx, :] = ni
                hf["results/te_frac"]  [wl_idx, :] = tf
                hf["results/n_guided"] [wl_idx]    = ng
                hf["flags/computed"]   [wl_idx]    = True
                hf.flush()

                runs_done += 1
                elapsed = time.time() - t_start
                rate    = runs_done / elapsed if elapsed > 0 else 1e-9
                eta     = (runs_total - runs_done) / rate
                log.info(
                    f"  [{runs_done:3d}/{runs_total}]  "
                    f"λ = {sweep_wl_nm[wl_idx]:.2f} nm  │  "
                    f"n_guided = {ng}  │  "
                    f"{rate:.2f} sim/s  │  ETA {eta:.0f} s"
                )

            hf["metadata"].attrs["timestamp_end"]  = datetime.now().isoformat()
            hf["metadata"].attrs["runs_completed"] = int(computed.sum())

    finally:
        mode.close()
        log.info("Lumerical MODE session closed.")

    total_t = time.time() - t_start
    log.info(
        f"Sweep done │ {runs_done} new runs │ "
        f"total = {total_t:.1f} s │ avg = {total_t/max(runs_done,1):.2f} s/sim"
    )

    return dict(neff_real=neff_real, neff_imag=neff_imag,
                te_frac=te_frac, n_guided=n_guided,
                computed=computed, sweep_wl_nm=sweep_wl_nm)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 3 — Execute sweep · Unpack · Post-process (ng, β₂, birefringence)    ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

# ── Run sweep (or load from HDF5 cache if already computed) ───────────────────
results = run_fde_wavelength_sweep()

neff_raw     = results["neff_real"]    # [N_wl, N_modes]
neff_imag    = results["neff_imag"]
te_frac_raw  = results["te_frac"]
n_guided_raw = results["n_guided"]
sweep_wl_nm  = results["sweep_wl_nm"]
sweep_wl_m   = sweep_wl_nm * 1e-9

# ─────────────────────────────────────────────────────────────────────────────
# Post-processing 1: Separate TE and TM fundamental modes
#
#   Lumerical orders modes by decreasing Re(neff). For SOI:
#     mode index 0  →  TE00  (highest neff, TE fraction ≈ 0.96–0.99)
#     mode index 1  →  TM00  (next, TE fraction ≈ 0.01–0.04)
#
#   We use TE threshold = 0.5 and guided criterion Re(neff) > N_SIO2_CUTOFF.
# ─────────────────────────────────────────────────────────────────────────────
TE_THRESHOLD = 0.5

def _separate_TE_TM(neff_2d: np.ndarray, te_2d: np.ndarray):
    """
    Separate fundamental TE and TM modes from the raw FDE output.
    Returns (neff_TE, neff_TM) as 1D arrays [N_wl], NaN where not found.
    """
    n_wl, n_mod = neff_2d.shape
    neff_TE = np.full(n_wl, np.nan)
    neff_TM = np.full(n_wl, np.nan)

    for wl_i in range(n_wl):
        te_found = tm_found = False
        for m_i in range(n_mod):
            ne = neff_2d[wl_i, m_i]
            tf = te_2d  [wl_i, m_i]
            if np.isnan(ne) or ne <= N_SIO2_CUTOFF:
                continue
            if tf > TE_THRESHOLD and not te_found:
                neff_TE[wl_i] = ne
                te_found = True
            elif tf <= TE_THRESHOLD and not tm_found:
                neff_TM[wl_i] = ne
                tm_found = True
            if te_found and tm_found:
                break

    return neff_TE, neff_TM

neff_TE, neff_TM = _separate_TE_TM(neff_raw, te_frac_raw)

# ─────────────────────────────────────────────────────────────────────────────
# Post-processing 2: Group index
#
#   ng = neff(λ) − λ · d(neff)/dλ                         (Eq. 1)
#
#   Uses np.gradient for central differences on the 51-point λ array.
#   2 nm step → accurate first derivative for ng and second for β₂.
# ─────────────────────────────────────────────────────────────────────────────
# First derivative of neff w.r.t. λ [1/m]
dneff_TE_dlam = np.gradient(neff_TE, sweep_wl_m)
dneff_TM_dlam = np.gradient(neff_TM, sweep_wl_m)

ng_TE = neff_TE - sweep_wl_m * dneff_TE_dlam    # group index TE
ng_TM = neff_TM - sweep_wl_m * dneff_TM_dlam    # group index TM

# ─────────────────────────────────────────────────────────────────────────────
# Post-processing 3: Group velocity dispersion β₂
#
#   β(ω) = neff(ω) · ω/c  →  β₂ = d²β/dω²
#
#   Working in wavelength domain:
#     β₂ = λ³ / (2πc²) · d²(neff)/dλ²              (Eq. 2)
#
#   Equivalently:  β₂ = −(λ²/2πc) · d(ng)/dλ        (Eq. 3)
#
#   Units: s²/m  →  convert to ps²/km  (1 s²/m = 10²⁷ ps²/km)
# ─────────────────────────────────────────────────────────────────────────────
d2neff_TE_dlam2 = np.gradient(dneff_TE_dlam, sweep_wl_m)   # second derivative
d2neff_TM_dlam2 = np.gradient(dneff_TM_dlam, sweep_wl_m)

# β₂ in s²/m
beta2_TE_s2m = sweep_wl_m**3 / (2.0 * np.pi * C_LIGHT_MS**2) * d2neff_TE_dlam2
beta2_TM_s2m = sweep_wl_m**3 / (2.0 * np.pi * C_LIGHT_MS**2) * d2neff_TM_dlam2

# Convert to ps²/km: 1 s²/m = 1e27 ps²/km
beta2_TE_ps2km = beta2_TE_s2m * 1e27
beta2_TM_ps2km = beta2_TM_s2m * 1e27

# Dispersion parameter D = −(2πc/λ²)·β₂  [ps/(nm·km)]
#   D > 0: anomalous dispersion  (β₂ < 0)
#   D < 0: normal dispersion     (β₂ > 0)
D_TE_ps_nm_km = -(2 * np.pi * C_LIGHT_MS / sweep_wl_m**2) * beta2_TE_s2m * 1e-3 * 1e9 * 1e12
# simplified: D [ps/(nm·km)] = β₂[s²/m] × (−2πc/λ²) × unit_factors
# Unit check:  D = -β₂ × 2πc/λ²  [s/m²] then × (10^12 ps/s) / (10^9 nm/m × 10^3 km/m)
# = -β₂ × 2πc/λ²  × 10^12 / (10^12) [ps/(nm·km)] — simplifies to:
D_TE_ps_nm_km = -(2.0 * np.pi * C_LIGHT_MS * 1e-12 / sweep_wl_m**2 / 1e-9 / 1e3) * beta2_TE_s2m
D_TM_ps_nm_km = -(2.0 * np.pi * C_LIGHT_MS * 1e-12 / sweep_wl_m**2 / 1e-9 / 1e3) * beta2_TM_s2m

# ─────────────────────────────────────────────────────────────────────────────
# Post-processing 4: Birefringence
#
#   B(λ) = neff_TE(λ) − neff_TM(λ)                        (Eq. 4)
#
#   In a 450 nm × 220 nm Si strip, the asymmetric cross-section (w/h ≈ 2.05)
#   and high index contrast (Δn ≈ 2.03) produce large birefringence B ≈ 0.6.
# ─────────────────────────────────────────────────────────────────────────────
birefringence = neff_TE - neff_TM

# ─────────────────────────────────────────────────────────────────────────────
# Save post-processed results back to HDF5 (new group: postprocessed/)
# ─────────────────────────────────────────────────────────────────────────────
with h5py.File(HDF5_PATH, "a") as hf:
    pp = hf.require_group("postprocessed")
    for name, data in [
        ("neff_TE",         neff_TE),
        ("neff_TM",         neff_TM),
        ("ng_TE",           ng_TE),
        ("ng_TM",           ng_TM),
        ("beta2_TE_s2m",    beta2_TE_s2m),
        ("beta2_TM_s2m",    beta2_TM_s2m),
        ("beta2_TE_ps2km",  beta2_TE_ps2km),
        ("beta2_TM_ps2km",  beta2_TM_ps2km),
        ("D_TE_ps_nm_km",   D_TE_ps_nm_km),
        ("birefringence",   birefringence),
    ]:
        if name in pp:
            pp[name][:] = data
        else:
            pp.create_dataset(name, data=data)
    hf.flush()
log.info(f"Post-processed results saved → {HDF5_PATH}")

# ─────────────────────────────────────────────────────────────────────────────
# Console summary table
# ─────────────────────────────────────────────────────────────────────────────
# Find index closest to 1550 nm
_idx1550 = int(np.argmin(np.abs(sweep_wl_nm - 1550.0)))

print("\n" + "=" * 80)
print("  POST-PROCESSING RESULTS — 220 nm SOI Strip Waveguide")
print(f"  {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  │  "
      f"Si / SiO₂  │  λ = {LAM_START_NM:.0f}–{LAM_STOP_NM:.0f} nm")
print("=" * 80)
print(f"  {'λ (nm)':>8}  {'neff_TE':>10}  {'neff_TM':>10}  "
      f"{'ng_TE':>9}  {'ng_TM':>9}  "
      f"{'β₂_TE (ps²/km)':>16}  {'B (birefr.)':>13}")
print("  " + "─" * 76)
for i, wl in enumerate(sweep_wl_nm):
    if i % 5 == 0 or i == _idx1550:
        marker = " ←" if i == _idx1550 else ""
        print(
            f"  {wl:>8.1f}  "
            f"{neff_TE[i]:>10.6f}  "
            f"{neff_TM[i]:>10.6f}  "
            f"{ng_TE[i]:>9.5f}  "
            f"{ng_TM[i]:>9.5f}  "
            f"{beta2_TE_ps2km[i]:>16.1f}  "
            f"{birefringence[i]:>13.6f}"
            f"{marker}"
        )
print("=" * 80)

# Zero-dispersion wavelength
_valid_TE = ~np.isnan(beta2_TE_ps2km)
if np.any(np.diff(np.sign(beta2_TE_ps2km[_valid_TE]))):
    _zero_idx = np.where(np.diff(np.sign(beta2_TE_ps2km[_valid_TE])))[0][0]
    ZDW_TE = float(sweep_wl_nm[_valid_TE][_zero_idx])
    print(f"\n  Zero-dispersion wavelength (TE00): λ_ZDW ≈ {ZDW_TE:.1f} nm")
else:
    print(f"\n  No zero-dispersion crossing found for TE in 1500–1600 nm range.")

print(f"\n  At λ = {sweep_wl_nm[_idx1550]:.0f} nm:")
print(f"    neff_TE  = {neff_TE[_idx1550]:.6f}   neff_TM  = {neff_TM[_idx1550]:.6f}")
print(f"    ng_TE    = {ng_TE[_idx1550]:.6f}     ng_TM    = {ng_TM[_idx1550]:.6f}")
print(f"    β₂_TE    = {beta2_TE_ps2km[_idx1550]:.2f} ps²/km  "
      f"  β₂_TM  = {beta2_TM_ps2km[_idx1550]:.2f} ps²/km")
print(f"    Birefr.  = {birefringence[_idx1550]:.6f}   (neff_TE − neff_TM)")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 4 — Publication-quality plots                                         ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.lines import Line2D

# ─────────────────────────────────────────────────────────────────────────────
# Plotting style
# ─────────────────────────────────────────────────────────────────────────────
plt.rcParams.update({
    "figure.dpi":        150,
    "font.family":       "DejaVu Sans",
    "font.size":         11,
    "axes.titlesize":    12,
    "axes.labelsize":    11,
    "xtick.labelsize":   9,
    "ytick.labelsize":   9,
    "legend.fontsize":   9,
    "lines.linewidth":   2.0,
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "axes.grid":         True,
    "grid.alpha":        0.30,
    "grid.linestyle":    "--",
})

SAVE_DPI  = 250
C_TE      = "#0072B2"    # blue   — TE mode
C_TM      = "#D55E00"    # orange — TM mode
C_NG      = "#009E73"    # green  — group index
C_ZERO    = "#555555"    # gray   — zero reference line


def _save(fig, stem: str) -> None:
    """Save figure as both PDF and PNG in DATA_DIR."""
    for ext in (".pdf", ".png"):
        out = str(DATA_DIR / (stem + ext))
        fig.savefig(out, dpi=SAVE_DPI, bbox_inches="tight")
        print(f"  Saved → {out}")


# ─────────────────────────────────────────────────────────────────────────────
# Figure 1: neff and ng vs wavelength  (TE and TM)
#
#   Two panels:
#     Left  — neff_TE and neff_TM vs λ
#     Right — ng_TE  and ng_TM  vs λ
#
#   Key observation: ng >> neff because the strong waveguide dispersion
#   (large |d(neff)/dλ|) adds a large negative correction: ng = neff − λ·d(neff)/dλ
# ─────────────────────────────────────────────────────────────────────────────
fig1, (ax1a, ax1b) = plt.subplots(1, 2, figsize=(14, 5.5))
fig1.suptitle(
    f"220 nm SOI Strip Waveguide — Modal Indices\n"
    f"Si core: {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  │  "
    f"SiO₂ cladding  │  dispersive material model",
    fontsize=12, fontweight="bold",
)

# Left: effective index
ax1a.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=2.2, label=r"$n_\mathrm{eff}$ (TE00)")
ax1a.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=2.2, ls="--", label=r"$n_\mathrm{eff}$ (TM00)")
ax1a.set_xlabel("Wavelength  λ  (nm)")
ax1a.set_ylabel(r"Effective index  $n_\mathrm{eff}$")
ax1a.set_title("Effective Index vs Wavelength")
ax1a.legend()
ax1a.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax1a.yaxis.set_minor_locator(ticker.AutoMinorLocator())
ax1a.annotate(
    r"$n_\mathrm{eff,TE} > n_\mathrm{eff,TM}$" + "\n(birefringence from\nasymmetric geometry)",
    xy=(1550, (neff_TE[_idx1550] + neff_TM[_idx1550]) / 2),
    xytext=(1530, 2.0),
    fontsize=8,
    arrowprops=dict(arrowstyle="->", color="gray", lw=0.8),
    color="gray",
)

# Right: group index
ax1b.plot(sweep_wl_nm, ng_TE, color=C_TE,  lw=2.2, label=r"$n_g$ (TE00)")
ax1b.plot(sweep_wl_nm, ng_TM, color=C_TM,  lw=2.2, ls="--", label=r"$n_g$ (TM00)")
ax1b.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=1.2, ls=":", alpha=0.55,
          label=r"$n_\mathrm{eff}$ (TE00, ref.)")
ax1b.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=1.2, ls=":",  alpha=0.55,
          label=r"$n_\mathrm{eff}$ (TM00, ref.)")
ax1b.set_xlabel("Wavelength  λ  (nm)")
ax1b.set_ylabel(r"Group index  $n_g = n_\mathrm{eff} - \lambda\,\mathrm{d}n_\mathrm{eff}/\mathrm{d}\lambda$")
ax1b.set_title(r"Group Index vs Wavelength  ($n_g > n_\mathrm{eff}$ due to strong WG dispersion)")
ax1b.legend(ncol=2, fontsize=8)
ax1b.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax1b.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig1.tight_layout()
_save(fig1, "HW4_SOI_neff_ng_vs_wavelength")
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# Figure 2: Group velocity dispersion β₂ vs wavelength
#
#   Also shows dispersion parameter D = −(2πc/λ²)·β₂  [ps/(nm·km)]
#   Key markers: zero-dispersion wavelength, 1550 nm operating point
# ─────────────────────────────────────────────────────────────────────────────
fig2, (ax2a, ax2b) = plt.subplots(1, 2, figsize=(14, 5.5))
fig2.suptitle(
    f"220 nm SOI Strip Waveguide — Group Velocity Dispersion\n"
    f"β₂ = λ³/(2πc²) · d²n_eff/dλ²     "
    f"Si {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  │  SiO₂ cladding",
    fontsize=12, fontweight="bold",
)

# Left: β₂ [ps²/km]
ax2a.plot(sweep_wl_nm, beta2_TE_ps2km, color=C_TE, lw=2.2, label=r"$\beta_2$ (TE00)")
ax2a.plot(sweep_wl_nm, beta2_TM_ps2km, color=C_TM, lw=2.2, ls="--", label=r"$\beta_2$ (TM00)")
ax2a.axhline(0, color=C_ZERO, lw=1.0, ls=":", alpha=0.8, label="β₂ = 0  (ZDW)")
ax2a.axvline(1550, color="gray", lw=0.8, ls=":", alpha=0.7, label="λ = 1550 nm")
ax2a.fill_between(sweep_wl_nm, beta2_TE_ps2km, 0,
                   where=(beta2_TE_ps2km < 0), alpha=0.10, color=C_TE,
                   label="Anomalous (β₂ < 0)")
ax2a.fill_between(sweep_wl_nm, beta2_TE_ps2km, 0,
                   where=(beta2_TE_ps2km >= 0), alpha=0.10, color=C_TM,
                   label="Normal (β₂ > 0)")
ax2a.set_xlabel("Wavelength  λ  (nm)")
ax2a.set_ylabel(r"$\beta_2$  (ps²/km)")
ax2a.set_title(r"GVD Coefficient  $\beta_2$  vs Wavelength")
ax2a.legend(ncol=2, fontsize=8)
ax2a.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax2a.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Right: D parameter [ps/(nm·km)]
ax2b.plot(sweep_wl_nm, D_TE_ps_nm_km, color=C_TE, lw=2.2, label=r"$D$ (TE00)")
ax2b.plot(sweep_wl_nm, D_TM_ps_nm_km, color=C_TM, lw=2.2, ls="--", label=r"$D$ (TM00)")
ax2b.axhline(0, color=C_ZERO, lw=1.0, ls=":", alpha=0.8)
ax2b.axvline(1550, color="gray", lw=0.8, ls=":", alpha=0.7)
ax2b.set_xlabel("Wavelength  λ  (nm)")
ax2b.set_ylabel(r"Dispersion parameter  $D$  (ps/(nm·km))")
ax2b.set_title(r"Dispersion $D = -(2\pi c/\lambda^2)\,\beta_2$  vs Wavelength")
ax2b.legend()
ax2b.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax2b.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig2.tight_layout()
_save(fig2, "HW4_SOI_GVD_vs_wavelength")
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# Figure 3: Birefringence  B(λ) = neff_TE − neff_TM
# ─────────────────────────────────────────────────────────────────────────────
fig3, ax3 = plt.subplots(figsize=(8, 5))
ax3.plot(sweep_wl_nm, birefringence, color="#009E73", lw=2.4,
         label=r"$B = n_\mathrm{eff,TE} - n_\mathrm{eff,TM}$")
ax3.axvline(1550, color="gray", lw=0.9, ls=":", alpha=0.8, label="λ = 1550 nm")
ax3.fill_between(sweep_wl_nm, birefringence, 0, alpha=0.12, color="#009E73")
ax3.set_xlabel("Wavelength  λ  (nm)")
ax3.set_ylabel(r"Modal birefringence  $B = n_\mathrm{eff,TE} - n_\mathrm{eff,TM}$")
ax3.set_title(
    f"Birefringence of 220 nm SOI Strip Waveguide\n"
    f"Si {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  │  "
    f"w/h ≈ {WG_WIDTH_NM/WG_HEIGHT_NM:.2f}  (asymmetric cross-section)"
)
ax3.legend()
ax3.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax3.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Annotate value at 1550 nm
ax3.annotate(
    f"B(1550 nm) = {birefringence[_idx1550]:.4f}",
    xy=(1550, birefringence[_idx1550]),
    xytext=(1560, birefringence[_idx1550] - 0.02),
    fontsize=9,
    arrowprops=dict(arrowstyle="->", color="#009E73", lw=1.0),
    color="#009E73",
)

fig3.tight_layout()
_save(fig3, "HW4_SOI_birefringence_vs_wavelength")
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# Figure 4: Summary 2×2 overview panel for the report
# ─────────────────────────────────────────────────────────────────────────────
fig4, axes4 = plt.subplots(2, 2, figsize=(13, 10))
fig4.suptitle(
    f"HW4 Part 1 — 220 nm SOI Platform Mode Analysis\n"
    f"Si {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm strip  │  "
    f"SiO₂ cladding  │  λ = {LAM_START_NM:.0f}–{LAM_STOP_NM:.0f} nm",
    fontsize=13, fontweight="bold",
)

# Panel (0,0): neff
ax = axes4[0, 0]
ax.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$n_\mathrm{eff}$")
ax.set_title("Effective Index"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Panel (0,1): ng
ax = axes4[0, 1]
ax.plot(sweep_wl_nm, ng_TE, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, ng_TM, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$n_g$")
ax.set_title("Group Index"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Panel (1,0): β₂
ax = axes4[1, 0]
ax.plot(sweep_wl_nm, beta2_TE_ps2km, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, beta2_TM_ps2km, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.axhline(0, color=C_ZERO, lw=0.9, ls=":", alpha=0.8)
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$\beta_2$ (ps²/km)")
ax.set_title(r"GVD Coefficient $\beta_2$"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Panel (1,1): birefringence
ax = axes4[1, 1]
ax.plot(sweep_wl_nm, birefringence, color="#009E73", lw=2.0,
        label=r"$B = n_\mathrm{eff,TE} - n_\mathrm{eff,TM}$")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"Birefringence $B$")
ax.set_title("Modal Birefringence"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig4.tight_layout()
_save(fig4, "HW4_SOI_overview_4panel")
plt.show()

print(f"\n  All figures saved to:  {DATA_DIR}")
print(f"  HDF5 data file:        {HDF5_PATH}")
```
