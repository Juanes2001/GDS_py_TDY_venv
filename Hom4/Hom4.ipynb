```python
# ═══════════════════════════════════════════════════════════════════════════════
# HW4 — Part 1 : 220 nm SOI Strip Waveguide — FDE Wavelength Sweep
# Version: Fixed-Index (constant n for Si and SiO₂ throughout sweep)
# Course : UNAL-BSU PIC Design — Spring 2025
# Student: Juan E. Rodriguez Ochoa
#
# Copy each cell (delimited by ╔═══╗ headers) into its own Jupyter notebook cell.
#
# CHANGES vs the original dispersive version:
#   1. FIXED REFRACTIVE INDICES for both core and cladding.
#      Instead of "Si (Silicon) - Palik" and "SiO2 (Glass) - Palik" (which
#      evaluate n(λ) at every wavelength step and caused the 1534 nm kink),
#      both materials are now set as "<Object defined dielectric>" with a
#      single constant scalar index.  The values chosen are the standard
#      references at λ = 1550 nm (mid-band):
#          N_SI_FIXED   = 3.476   (Si core)
#          N_SIO2_FIXED = 1.444   (SiO₂ cladding everywhere)
#      These values are constant across the full 1500–1600 nm sweep.
#
#   2. CLADDING GEOMETRY CONFIRMED + DOCUMENTED.
#      The original code used a single "SiO2_bg" rectangle that spans the
#      entire FDE domain (6.45 µm wide × 5.22 µm tall).  The "Si_core"
#      rectangle (450 nm × 220 nm) is added on top and overrides only its
#      own area.  This correctly places SiO₂:
#        • ABOVE the core  (+110 nm → +2 µm, 2.0 µm top cladding)
#        • BELOW the core  (−110 nm → −3.11 µm, 3.0 µm BOX)
#        • LEFT  of core   (−225 nm → −3.225 µm, 3.0 µm margin)
#        • RIGHT of core   (+225 nm → +3.225 µm, 3.0 µm margin)
#      No change to the geometry was needed; this note confirms correctness.
#
#   3. FIGURE 2 LAYOUT — β₂ and D plots merged (1×2 instead of 2×2):
#      Left panel  : β₂  for TE00 and TM00 on the same axes.
#      Right panel : D   for TE00 and TM00 on the same axes.
# ═══════════════════════════════════════════════════════════════════════════════


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 1 — Imports · Lumerical API path · Global parameters                 ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

import sys
import os
import platform
import time
import logging
from pathlib import Path
from datetime import datetime

import numpy as np
import h5py

# ─────────────────────────────────────────────────────────────────────────────
# Lumerical installation path
# Set LUMERICAL_VERSION to match the folder name under C:\Program Files\Lumerical\
# ─────────────────────────────────────────────────────────────────────────────
LUMERICAL_VERSION = "v202"          # ← update if your folder name differs

if platform.system() == "Windows":
    LUMERICAL_ROOT = rf"C:\Program Files\Lumerical\{LUMERICAL_VERSION}"
    LUMERICAL_API  = rf"{LUMERICAL_ROOT}\api\python"
    LUMERICAL_BIN  = rf"{LUMERICAL_ROOT}\bin"
else:
    LUMERICAL_ROOT = f"/opt/lumerical/{LUMERICAL_VERSION}"
    LUMERICAL_API  = f"{LUMERICAL_ROOT}/api/python"
    LUMERICAL_BIN  = f"{LUMERICAL_ROOT}/bin"

if "lumapi" in sys.modules:
    del sys.modules["lumapi"]

if LUMERICAL_API not in sys.path:
    sys.path.insert(0, LUMERICAL_API)

if platform.system() == "Windows" and hasattr(os, "add_dll_directory"):
    os.add_dll_directory(str(LUMERICAL_BIN))

assert Path(LUMERICAL_API).exists(), (
    f"Lumerical API not found:\n  {LUMERICAL_API}\n"
    f"  Check LUMERICAL_VERSION = '{LUMERICAL_VERSION}'"
)
assert Path(LUMERICAL_BIN).exists(), f"Lumerical bin not found:\n  {LUMERICAL_BIN}"

import lumapi  # noqa — must come after all path setup
print(f"lumapi imported successfully from:\n  {lumapi.__file__}")

# ─────────────────────────────────────────────────────────────────────────────
# Logging
# ─────────────────────────────────────────────────────────────────────────────
logging.basicConfig(
    level   = logging.INFO,
    format  = "%(asctime)s │ %(levelname)s │ %(message)s",
    datefmt = "%H:%M:%S",
)
log = logging.getLogger("HW4_SOI_FDE")

# ─────────────────────────────────────────────────────────────────────────────
# I/O paths
# New VERSION_NAME avoids accidentally loading the old dispersive-model cache.
# ─────────────────────────────────────────────────────────────────────────────
VERSION_NAME = "HW4_SOI_220nm_strip_FDE_fixed_index"
DATA_DIR     = Path.cwd() / "data_HW4_SOI_mode_analysis"
DATA_DIR.mkdir(parents=True, exist_ok=True)
HDF5_PATH    = DATA_DIR / f"{VERSION_NAME}.h5"

# ─────────────────────────────────────────────────────────────────────────────
# 220 nm SOI platform — waveguide cross-section
#
#   Full cross-section in the Y-Z plane (X = propagation direction):
#
#   z [µm]
#   +2.11 ┌──────────────────────────────────────────────────────────┐
#         │  SiO₂  (top cladding, 2.00 µm)                          │
#   +0.11 │          ┌──────────────────┐                            │
#         │  SiO₂   │   Si core         │  SiO₂                    │
#         │  (left  │  450 nm × 220 nm  │  (right cladding,        │
#         │  3.0 µm)│                   │   3.0 µm)                │
#  -0.11  │          └──────────────────┘                            │
#         │  SiO₂  (BOX / buried oxide, 3.00 µm)                    │
#  -3.11  └──────────────────────────────────────────────────────────┘
#         y: -3.225 µm ←────────────────────────────→ +3.225 µm
#
#   SiO₂ is present on ALL FOUR SIDES of the Si core:
#     Top    : +110 nm → +2110 nm   (2.00 µm, SIM_Z_ABOVE_UM)
#     Bottom : -110 nm → -3110 nm   (3.00 µm, SIM_Z_BELOW_UM = BOX)
#     Left   : -225 nm → -3225 nm   (3.00 µm, SIM_Y_MARGIN_UM)
#     Right  : +225 nm → +3225 nm   (3.00 µm, SIM_Y_MARGIN_UM)
#
#   Implementation: one "SiO2_bg" rectangle fills the entire domain; the
#   "Si_core" rectangle overrides only the 450×220 nm central region.
#   Lumerical gives higher priority to the later-added object (Si_core),
#   so SiO₂ automatically fills all four surrounding areas.
# ─────────────────────────────────────────────────────────────────────────────
WG_WIDTH_NM  = 450.0     # strip width                 [nm]
WG_HEIGHT_NM = 220.0     # SOI device layer thickness  [nm]

SIM_Z_BELOW_UM  = 3.0    # SiO₂ BOX depth below core  [µm]
SIM_Z_ABOVE_UM  = 2.0    # SiO₂ top cladding above     [µm]
SIM_Y_MARGIN_UM = 3.0    # SiO₂ margin each lateral side [µm]

N_MODES_REQUEST = 4

# ─────────────────────────────────────────────────────────────────────────────
# FIXED refractive indices — constant across the full 1500–1600 nm sweep
#
#   Standard values at λ = 1550 nm (mid-band reference):
#     Si   : n = 3.476  (Palik / Green 1995)
#     SiO₂ : n = 1.444  (Palik / Malitson 1965)
#
#   Using constant indices removes all wavelength-dependent material effects.
#   The neff(λ) variation in the sweep then reflects ONLY the waveguide
#   geometric dispersion (how mode confinement changes with λ at fixed n).
#   This is a clean and transparent approach for studying waveguide dispersion.
#
#   In Lumerical these are set via the "<Object defined dielectric>" material
#   type, which accepts a single real scalar index.
# ─────────────────────────────────────────────────────────────────────────────
N_SI_FIXED   = 3.476    # Si core refractive index   (constant)
N_SIO2_FIXED = 1.444    # SiO₂ cladding index        (constant)

# Guided-mode cutoff: modes with Re(neff) ≤ N_SIO2_FIXED are radiation modes.
# Set slightly below the cladding index to keep the threshold sharp.
N_SIO2_CUTOFF = N_SIO2_FIXED - 0.005   # = 1.439

# ─────────────────────────────────────────────────────────────────────────────
# Wavelength sweep: 1500–1600 nm, 51 uniformly-spaced points (Δλ = 2 nm)
# ─────────────────────────────────────────────────────────────────────────────
LAM_START_NM  = 1500.0
LAM_STOP_NM   = 1600.0
N_WAVELENGTHS = 51
SWEEP_WL_NM   = np.linspace(LAM_START_NM, LAM_STOP_NM, N_WAVELENGTHS)  # [nm]
SWEEP_WL_M    = SWEEP_WL_NM * 1e-9                                       # [m]
SWEEP_WL_UM   = SWEEP_WL_NM * 1e-3                                       # [µm]

C_LIGHT_MS = 2.99792458e8   # speed of light [m/s]

# ─────────────────────────────────────────────────────────────────────────────
# Derived geometry [µm]
# ─────────────────────────────────────────────────────────────────────────────
WG_W_UM     = WG_WIDTH_NM  * 1e-3     # 0.450 µm
WG_H_UM     = WG_HEIGHT_NM * 1e-3     # 0.220 µm
_half_h_um  = WG_H_UM / 2.0           # 0.110 µm

# Simulation domain total extents
SIM_Y_SPAN_UM = WG_W_UM + 2.0 * SIM_Y_MARGIN_UM         # 6.450 µm (y)
SIM_Z_SPAN_UM = SIM_Z_BELOW_UM + WG_H_UM + SIM_Z_ABOVE_UM  # 5.220 µm (z)

# Domain centre in z — offset because BOX (3 µm) ≠ top cladding (2 µm)
SIM_Z_CTR_UM = (SIM_Z_ABOVE_UM - SIM_Z_BELOW_UM) / 2.0   # −0.500 µm

# ─────────────────────────────────────────────────────────────────────────────
# Mesh sizing: ≥ 15 grid cells per wavelength inside Si
#   step ≈ λ_min / (N_SI_FIXED × 15) = 1500 nm / (3.476 × 15) ≈ 28.8 nm
# ─────────────────────────────────────────────────────────────────────────────
MESH_CELLS_PER_WVL = 15
_mesh_step_um      = SWEEP_WL_UM.min() / (N_SI_FIXED * MESH_CELLS_PER_WVL)
MESH_CELLS_Y       = int(np.ceil(SIM_Y_SPAN_UM / _mesh_step_um))
MESH_CELLS_Z       = int(np.ceil(SIM_Z_SPAN_UM / _mesh_step_um))

# ─────────────────────────────────────────────────────────────────────────────
# Print configuration summary
# ─────────────────────────────────────────────────────────────────────────────
print("=" * 65)
print("  220 nm SOI Strip Waveguide — FDE Wavelength Sweep")
print("  Material model: FIXED REFRACTIVE INDEX (constant n)")
print("=" * 65)
print(f"  Si  core    : n = {N_SI_FIXED}  (constant, ref. λ = 1550 nm)")
print(f"  SiO₂ cladding: n = {N_SIO2_FIXED}  (constant, all four sides)")
print(f"  Waveguide   : {WG_WIDTH_NM:.0f} nm × {WG_HEIGHT_NM:.0f} nm  (w × h)")
print(f"  Cladding    : TOP {SIM_Z_ABOVE_UM:.1f} µm  │  "
      f"BOX {SIM_Z_BELOW_UM:.1f} µm  │  SIDES ±{SIM_Y_MARGIN_UM:.1f} µm")
print(f"  λ sweep     : {LAM_START_NM:.0f}–{LAM_STOP_NM:.0f} nm  "
      f"({N_WAVELENGTHS} pts, Δλ = 2 nm)")
print(f"  Domain Y    : {SIM_Y_SPAN_UM:.3f} µm  ({MESH_CELLS_Y} cells)")
print(f"  Domain Z    : {SIM_Z_SPAN_UM:.3f} µm  ({MESH_CELLS_Z} cells)")
print(f"  Mesh step   : ≈ {_mesh_step_um*1e3:.1f} nm  "
      f"(≥ {MESH_CELLS_PER_WVL} cells/λ in Si)")
print(f"  neff cutoff : {N_SIO2_CUTOFF:.3f}  (just below n_SiO₂ = {N_SIO2_FIXED})")
print(f"  HDF5        : {HDF5_PATH}")
print("=" * 65)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 2 — FDE geometry builder · mode extractor · HDF5 init · sweep        ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

def _build_fde_SOI(
    mode,
    wavelength_m    : float,
    wg_w_m          : float = WG_WIDTH_NM  * 1e-9,
    wg_h_m          : float = WG_HEIGHT_NM * 1e-9,
    sim_y_span_um   : float = SIM_Y_SPAN_UM,
    sim_z_span_um   : float = SIM_Z_SPAN_UM,
    sim_z_ctr_um    : float = SIM_Z_CTR_UM,
    mesh_cells_y    : int   = MESH_CELLS_Y,
    mesh_cells_z    : int   = MESH_CELLS_Z,
    n_modes_request : int   = N_MODES_REQUEST,
    n_si            : float = N_SI_FIXED,
    n_sio2          : float = N_SIO2_FIXED,
) -> None:
    """
    Build the 220 nm SOI FDE geometry using fixed scalar refractive indices.

    Two rectangles:
      SiO2_bg  — spans full domain, n = N_SIO2_FIXED (SiO₂ on all 4 sides).
      Si_core  — 450 nm × 220 nm at origin, n = N_SI_FIXED, overrides SiO2_bg.
    """
    m = mode
    m.switchtolayout()
    m.selectall()
    m.delete()

    # ── FDE solver region ────────────────────────────────────────────────────
    m.addfde()
    m.set("solver type",           "2D X normal")
    m.set("x",                     0.0)
    m.set("y",                     0.0)
    m.set("z",                     sim_z_ctr_um * 1e-6)
    m.set("y span",                sim_y_span_um * 1e-6)
    m.set("z span",                sim_z_span_um * 1e-6)
    m.set("wavelength",            wavelength_m)
    m.set("number of trial modes", n_modes_request)
    m.set("mesh cells y",          mesh_cells_y)
    m.set("mesh cells z",          mesh_cells_z)

    # ── Rectangle 1: SiO₂ background — fills full domain (all 4 sides) ───────
    m.addrect()
    m.set("name",     "SiO2_bg")
    m.set("x",        0.0);  m.set("x span",  1.0e-6)
    m.set("y",        0.0);  m.set("y span",  sim_y_span_um * 1e-6)
    m.set("z",        sim_z_ctr_um * 1e-6)
    m.set("z span",   sim_z_span_um * 1e-6)
    m.set("material", "<Object defined dielectric>")
    m.set("index",    n_sio2)                           # N_SIO2_FIXED = 1.444

    # ── Rectangle 2: Si waveguide core — overrides SiO₂ in core area ─────────
    m.addrect()
    m.set("name",     "Si_core")
    m.set("x",        0.0);  m.set("x span",  1.0e-6)
    m.set("y",        0.0);  m.set("y span",  wg_w_m)
    m.set("z",        0.0);  m.set("z span",  wg_h_m)
    m.set("material", "<Object defined dielectric>")
    m.set("index",    n_si)                             # N_SI_FIXED = 3.476


def _extract_modes_SOI(mode, n_modes_request: int = N_MODES_REQUEST):
    """
    Extract neff and TE fraction for up to n_modes_request modes from the FDE.

    Returns
    -------
    neff_real : (n_modes_request,)  Re(neff),  NaN if mode absent
    neff_imag : (n_modes_request,)  Im(neff)
    te_frac   : (n_modes_request,)  TE polarisation fraction [0, 1]
    n_guided  : int                 count of guided modes found
    """
    neff_real = np.full(n_modes_request, np.nan, dtype=np.float64)
    neff_imag = np.full(n_modes_request, np.nan, dtype=np.float64)
    te_frac   = np.full(n_modes_request, np.nan, dtype=np.float64)
    n_guided  = 0

    for m_i in range(1, n_modes_request + 1):
        grp = f"FDE::data::mode{m_i}"
        try:
            raw_neff = mode.getdata(grp, "neff")
            raw_te   = mode.getdata(grp, "TE polarization fraction")
        except Exception:
            break

        neff_c = complex(np.asarray(raw_neff).flat[0])
        te_v   = float(np.asarray(raw_te).flat[0])

        neff_real[m_i - 1] = neff_c.real
        neff_imag[m_i - 1] = neff_c.imag
        te_frac  [m_i - 1] = te_v

        if neff_c.real > N_SIO2_CUTOFF:
            n_guided += 1

    return neff_real, neff_imag, te_frac, n_guided


def _init_hdf5_SOI(path: Path, n_wl: int, n_modes: int,
                   sweep_wl_nm: np.ndarray, metadata: dict) -> None:
    with h5py.File(path, "w") as f:
        md = f.create_group("metadata")
        md.create_dataset("sweep_wavelengths_nm", data=sweep_wl_nm)
        md.create_dataset("sweep_wavelengths_m",  data=sweep_wl_nm * 1e-9)
        for k, v in metadata.items():
            md.attrs[k] = v
        md.attrs["timestamp_start"] = datetime.now().isoformat()

        nan2d = np.full((n_wl, n_modes), np.nan, dtype=np.float64)
        chunk = (1, n_modes)
        rg = f.create_group("results")
        rg.create_dataset("neff_real", data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("neff_imag", data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("te_frac",   data=nan2d.copy(), chunks=chunk)
        rg.create_dataset("n_guided",
                          data=np.zeros(n_wl, dtype=np.int32), chunks=(1,))

        f.create_group("flags").create_dataset(
            "computed", data=np.zeros(n_wl, dtype=bool), chunks=(1,)
        )


def run_fde_wavelength_sweep(
    sweep_wl_nm  = SWEEP_WL_NM,
    hdf5_path    = HDF5_PATH,
    version_name = VERSION_NAME,
    hide_gui     = False,
) -> dict:
    """
    2D FDE wavelength sweep — fixed refractive indices for Si and SiO₂.

    At each wavelength:
      1. Updates FDE wavelength (geometry/material unchanged — fixed index).
      2. Runs mesh → findmodes().
      3. Extracts neff (real+imag) and TE fraction for all modes.
      4. Writes to HDF5 immediately (fault-safe caching).

    Re-running skips already-computed wavelength points.
    """
    n_wl  = len(sweep_wl_nm)
    n_mod = N_MODES_REQUEST
    sweep_wl_m = sweep_wl_nm * 1e-9

    neff_real = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    neff_imag = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    te_frac   = np.full((n_wl, n_mod), np.nan, dtype=np.float64)
    n_guided  = np.zeros(n_wl, dtype=np.int32)
    computed  = np.zeros(n_wl, dtype=bool)

    if hdf5_path.exists():
        log.info(f"Cache found → {hdf5_path}")
        with h5py.File(hdf5_path, "r") as f:
            if "results/neff_real" in f:
                neff_real[:] = f["results/neff_real"][:]
                neff_imag[:] = f["results/neff_imag"][:]
                te_frac[:]   = f["results/te_frac"][:]
                n_guided[:]  = f["results/n_guided"][:]
                computed[:]  = f["flags/computed"][:]
        n_cached = int(computed.sum()); remaining = n_wl - n_cached
        log.info(f"Cached: {n_cached}/{n_wl}  |  Remaining: {remaining}")
        if remaining == 0:
            log.info("All points cached — returning results.")
            return dict(neff_real=neff_real, neff_imag=neff_imag,
                        te_frac=te_frac, n_guided=n_guided,
                        computed=computed, sweep_wl_nm=sweep_wl_nm)
    else:
        log.info("No cache — initialising new HDF5 file.")
        _init_hdf5_SOI(
            path        = hdf5_path,
            n_wl        = n_wl,
            n_modes     = n_mod,
            sweep_wl_nm = sweep_wl_nm,
            metadata    = {
                "version_name"    : version_name,
                "material_model"  : "fixed_index",
                "n_si_fixed"      : N_SI_FIXED,
                "n_sio2_fixed"    : N_SIO2_FIXED,
                "wg_width_nm"     : WG_WIDTH_NM,
                "wg_height_nm"    : WG_HEIGHT_NM,
                "sim_z_below_um"  : SIM_Z_BELOW_UM,
                "sim_z_above_um"  : SIM_Z_ABOVE_UM,
                "sim_y_margin_um" : SIM_Y_MARGIN_UM,
                "mesh_cells_y"    : MESH_CELLS_Y,
                "mesh_cells_z"    : MESH_CELLS_Z,
                "n_modes_request" : n_mod,
                "n_sio2_cutoff"   : N_SIO2_CUTOFF,
            },
        )

    log.info(f"Launching Lumerical MODE …  "
             f"(n_Si={N_SI_FIXED}, n_SiO₂={N_SIO2_FIXED}, fixed)")
    mode       = lumapi.MODE(hide=hide_gui)
    t_start    = time.time()
    runs_done  = 0
    runs_total = int((~computed).sum())

    try:
        first_uncached = int(np.where(~computed)[0][0])
        _build_fde_SOI(mode, wavelength_m=float(sweep_wl_m[first_uncached]))
        log.info(f"Geometry built  (n_Si={N_SI_FIXED}, n_SiO₂={N_SIO2_FIXED})  "
                 f"│  {runs_total} points remaining …")

        with h5py.File(hdf5_path, "r+") as hf:
            for wl_idx, wl_m in enumerate(sweep_wl_m):
                if computed[wl_idx]:
                    continue

                mode.switchtolayout()
                mode.setnamed("FDE", "wavelength", float(wl_m))

                try:
                    mode.run()
                    mode.findmodes()
                except Exception as exc:
                    log.warning(
                        f"  FDE FAILED │ λ = {sweep_wl_nm[wl_idx]:.2f} nm │ {exc}"
                    )
                    computed[wl_idx]             = True
                    hf["flags/computed"][wl_idx] = True
                    hf.flush()
                    continue

                nr, ni, tf, ng = _extract_modes_SOI(mode, n_mod)

                neff_real[wl_idx, :] = nr
                neff_imag[wl_idx, :] = ni
                te_frac  [wl_idx, :] = tf
                n_guided [wl_idx]    = ng
                computed [wl_idx]    = True

                hf["results/neff_real"][wl_idx, :] = nr
                hf["results/neff_imag"][wl_idx, :] = ni
                hf["results/te_frac"]  [wl_idx, :] = tf
                hf["results/n_guided"] [wl_idx]    = ng
                hf["flags/computed"]   [wl_idx]    = True
                hf.flush()

                runs_done += 1
                elapsed = time.time() - t_start
                rate    = runs_done / elapsed if elapsed > 0 else 1e-9
                eta     = (runs_total - runs_done) / rate
                log.info(
                    f"  [{runs_done:3d}/{runs_total}]  "
                    f"λ = {sweep_wl_nm[wl_idx]:.2f} nm  │  "
                    f"neff_m0 = {nr[0]:.6f}  │  "
                    f"n_guided = {ng}  │  ETA {eta:.0f} s"
                )

            hf["metadata"].attrs["timestamp_end"]  = datetime.now().isoformat()
            hf["metadata"].attrs["runs_completed"] = int(computed.sum())

    finally:
        mode.close()
        log.info("Lumerical MODE session closed.")

    elapsed_total = time.time() - t_start
    log.info(
        f"Sweep done │ {runs_done} new runs │ "
        f"total = {elapsed_total:.1f} s │ avg = {elapsed_total/max(runs_done,1):.2f} s/sim"
    )

    return dict(neff_real=neff_real, neff_imag=neff_imag,
                te_frac=te_frac, n_guided=n_guided,
                computed=computed, sweep_wl_nm=sweep_wl_nm)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 3 — Execute sweep · Unpack · Post-process (ng, β₂, birefringence)    ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

results     = run_fde_wavelength_sweep()
neff_raw    = results["neff_real"]
neff_imag   = results["neff_imag"]
te_frac_raw = results["te_frac"]
sweep_wl_nm = results["sweep_wl_nm"]
sweep_wl_m  = sweep_wl_nm * 1e-9

# ─────────────────────────────────────────────────────────────────────────────
# Separate TE00 and TM00
# ─────────────────────────────────────────────────────────────────────────────
TE_THRESHOLD = 0.5

def _separate_TE_TM(neff_2d, te_2d):
    n_wl, n_mod = neff_2d.shape
    neff_TE = np.full(n_wl, np.nan)
    neff_TM = np.full(n_wl, np.nan)
    for wl_i in range(n_wl):
        te_found = tm_found = False
        for m_i in range(n_mod):
            ne = neff_2d[wl_i, m_i]; tf = te_2d[wl_i, m_i]
            if np.isnan(ne) or ne <= N_SIO2_CUTOFF:
                continue
            if tf > TE_THRESHOLD and not te_found:
                neff_TE[wl_i] = ne; te_found = True
            elif tf <= TE_THRESHOLD and not tm_found:
                neff_TM[wl_i] = ne; tm_found = True
            if te_found and tm_found:
                break
    return neff_TE, neff_TM

neff_TE, neff_TM = _separate_TE_TM(neff_raw, te_frac_raw)

# ─────────────────────────────────────────────────────────────────────────────
# Group index:  ng = neff − λ · d(neff)/dλ
# ─────────────────────────────────────────────────────────────────────────────
dneff_TE_dlam = np.gradient(neff_TE, sweep_wl_m)
dneff_TM_dlam = np.gradient(neff_TM, sweep_wl_m)

ng_TE = neff_TE - sweep_wl_m * dneff_TE_dlam
ng_TM = neff_TM - sweep_wl_m * dneff_TM_dlam

# ─────────────────────────────────────────────────────────────────────────────
# GVD:  β₂ = λ³/(2πc²) · d²(neff)/dλ²
# ─────────────────────────────────────────────────────────────────────────────
d2neff_TE_dlam2 = np.gradient(dneff_TE_dlam, sweep_wl_m)
d2neff_TM_dlam2 = np.gradient(dneff_TM_dlam, sweep_wl_m)

beta2_TE_s2m   = (sweep_wl_m**3 / (2.0 * np.pi * C_LIGHT_MS**2)) * d2neff_TE_dlam2
beta2_TM_s2m   = (sweep_wl_m**3 / (2.0 * np.pi * C_LIGHT_MS**2)) * d2neff_TM_dlam2
beta2_TE_ps2km = beta2_TE_s2m * 1e27
beta2_TM_ps2km = beta2_TM_s2m * 1e27

# ─────────────────────────────────────────────────────────────────────────────
# Dispersion parameter:  D [ps/(nm·km)] = −(2πc/λ²) · β₂[s²/m] · 10⁶
# ─────────────────────────────────────────────────────────────────────────────
D_TE_ps_nm_km = -(2.0 * np.pi * C_LIGHT_MS / sweep_wl_m**2) * beta2_TE_s2m * 1e6
D_TM_ps_nm_km = -(2.0 * np.pi * C_LIGHT_MS / sweep_wl_m**2) * beta2_TM_s2m * 1e6

# ─────────────────────────────────────────────────────────────────────────────
# Birefringence:  B(λ) = neff_TE − neff_TM
# ─────────────────────────────────────────────────────────────────────────────
birefringence = neff_TE - neff_TM

# ─────────────────────────────────────────────────────────────────────────────
# Save to HDF5
# ─────────────────────────────────────────────────────────────────────────────
with h5py.File(HDF5_PATH, "a") as hf:
    pp = hf.require_group("postprocessed")
    pp.attrs["derivative_method"] = "np.gradient (central differences)"
    pp.attrs["material_model"]    = "fixed_index"
    pp.attrs["n_si"]              = N_SI_FIXED
    pp.attrs["n_sio2"]            = N_SIO2_FIXED
    for name, data in [
        ("neff_TE",        neff_TE),
        ("neff_TM",        neff_TM),
        ("ng_TE",          ng_TE),
        ("ng_TM",          ng_TM),
        ("beta2_TE_s2m",   beta2_TE_s2m),
        ("beta2_TM_s2m",   beta2_TM_s2m),
        ("beta2_TE_ps2km", beta2_TE_ps2km),
        ("beta2_TM_ps2km", beta2_TM_ps2km),
        ("D_TE_ps_nm_km",  D_TE_ps_nm_km),
        ("D_TM_ps_nm_km",  D_TM_ps_nm_km),
        ("birefringence",  birefringence),
    ]:
        if name in pp: del pp[name]
        pp.create_dataset(name, data=data)
    hf.flush()
log.info(f"Post-processed results saved → {HDF5_PATH}")

# ─────────────────────────────────────────────────────────────────────────────
# Console summary
# ─────────────────────────────────────────────────────────────────────────────
_idx1550 = int(np.argmin(np.abs(sweep_wl_nm - 1550.0)))

print("\n" + "=" * 82)
print(f"  RESULTS — 220 nm SOI  │  {WG_WIDTH_NM:.0f}×{WG_HEIGHT_NM:.0f} nm  │  "
      f"n_Si={N_SI_FIXED}  n_SiO₂={N_SIO2_FIXED}  (fixed)")
print("=" * 82)
print(f"  {'λ (nm)':>8}  {'neff_TE':>10}  {'neff_TM':>10}  "
      f"{'ng_TE':>9}  {'ng_TM':>9}  "
      f"{'β₂_TE (ps²/km)':>16}  {'B':>10}")
print("  " + "─" * 78)
for i, wl in enumerate(sweep_wl_nm):
    if i % 5 == 0 or i == _idx1550:
        marker = "  ← 1550 nm" if i == _idx1550 else ""
        print(
            f"  {wl:>8.1f}  "
            f"{neff_TE[i]:>10.6f}  "
            f"{neff_TM[i]:>10.6f}  "
            f"{ng_TE[i]:>9.5f}  "
            f"{ng_TM[i]:>9.5f}  "
            f"{beta2_TE_ps2km[i]:>16.2f}  "
            f"{birefringence[i]:>10.6f}"
            f"{marker}"
        )
print("=" * 82)

_valid_TE = ~np.isnan(beta2_TE_ps2km)
_b2_v = beta2_TE_ps2km[_valid_TE]; _wl_v = sweep_wl_nm[_valid_TE]
_cross = np.where(np.diff(np.sign(_b2_v)) != 0)[0]
if len(_cross) > 0:
    ci = _cross[0]
    ZDW_TE = float(_wl_v[ci] - _b2_v[ci] *
                   (_wl_v[ci+1]-_wl_v[ci]) / (_b2_v[ci+1]-_b2_v[ci]))
    print(f"\n  ZDW (TE00) ≈ {ZDW_TE:.1f} nm")
else:
    ZDW_TE = None
    print(f"\n  ZDW (TE00): not found in {LAM_START_NM:.0f}–{LAM_STOP_NM:.0f} nm range")

print(f"\n  At λ = 1550 nm:")
print(f"    neff_TE  = {neff_TE[_idx1550]:.6f}   neff_TM  = {neff_TM[_idx1550]:.6f}")
print(f"    ng_TE    = {ng_TE[_idx1550]:.6f}     ng_TM    = {ng_TM[_idx1550]:.6f}")
print(f"    β₂_TE   = {beta2_TE_ps2km[_idx1550]:.2f} ps²/km")
print(f"    D_TE     = {D_TE_ps_nm_km[_idx1550]:.2f} ps/(nm·km)")
print(f"    B        = {birefringence[_idx1550]:.6f}  (neff_TE − neff_TM)")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  CELL 4 — Publication-quality plots                                         ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.rcParams.update({
    "figure.dpi":        150,
    "font.family":       "DejaVu Sans",
    "font.size":         11,
    "axes.titlesize":    12,
    "axes.labelsize":    11,
    "xtick.labelsize":   9,
    "ytick.labelsize":   9,
    "legend.fontsize":   9,
    "lines.linewidth":   2.0,
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "axes.grid":         True,
    "grid.alpha":        0.30,
    "grid.linestyle":    "--",
})

SAVE_DPI = 250
C_TE  = "#0072B2"
C_TM  = "#D55E00"
C_ZERO = "#555555"
C_ZDW  = "#009E73"

_subtitle = (f"Si {WG_WIDTH_NM:.0f}×{WG_HEIGHT_NM:.0f} nm  │  "
             f"n_Si={N_SI_FIXED}  n_SiO₂={N_SIO2_FIXED}  (fixed index)  │  "
             f"SiO₂ cladding on all 4 sides")

def _save(fig, stem: str) -> None:
    for ext in (".pdf", ".png"):
        out = str(DATA_DIR / (stem + ext))
        fig.savefig(out, dpi=SAVE_DPI, bbox_inches="tight")
        print(f"  Saved → {out}")


# ── Figure 1: neff and ng vs wavelength ───────────────────────────────────────
fig1, (ax1a, ax1b) = plt.subplots(1, 2, figsize=(14, 5.5))
fig1.suptitle("220 nm SOI Strip Waveguide — Modal Indices\n" + _subtitle,
              fontsize=12, fontweight="bold")

ax1a.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=2.2,
          label=r"$n_\mathrm{eff}$ (TE00)")
ax1a.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=2.2, ls="--",
          label=r"$n_\mathrm{eff}$ (TM00)")
ax1a.set_xlabel("Wavelength  λ  (nm)")
ax1a.set_ylabel(r"Effective index  $n_\mathrm{eff}$")
ax1a.set_title("Effective Index vs Wavelength")
ax1a.legend()
ax1a.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax1a.yaxis.set_minor_locator(ticker.AutoMinorLocator())

ax1b.plot(sweep_wl_nm, ng_TE,   color=C_TE, lw=2.2, label=r"$n_g$ (TE00)")
ax1b.plot(sweep_wl_nm, ng_TM,   color=C_TM, lw=2.2, ls="--", label=r"$n_g$ (TM00)")
ax1b.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=1.2, ls=":", alpha=0.50,
          label=r"$n_\mathrm{eff}$ TE00 (ref.)")
ax1b.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=1.2, ls=":", alpha=0.50,
          label=r"$n_\mathrm{eff}$ TM00 (ref.)")
ax1b.set_xlabel("Wavelength  λ  (nm)")
ax1b.set_ylabel(r"Group index  "
                r"$n_g = n_\mathrm{eff} - \lambda\,\mathrm{d}n_\mathrm{eff}/\mathrm{d}\lambda$")
ax1b.set_title(r"Group Index vs Wavelength  ($n_g > n_\mathrm{eff}$: WG dispersion)")
ax1b.legend(ncol=2, fontsize=8)
ax1b.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax1b.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig1.tight_layout()
_save(fig1, "HW4_SOI_neff_ng_vs_wavelength")
plt.show()


# ── Figure 2: β₂ (left, TE+TM) and D (right, TE+TM) — 1×2 layout ────────────
#
#   Left panel  — β₂ for TE00 and TM00 on the same axes.
#   Right panel — D  for TE00 and TM00 on the same axes.
# ─────────────────────────────────────────────────────────────────────────────
fig2, (ax2a, ax2b) = plt.subplots(1, 2, figsize=(14, 5.5))
fig2.suptitle(
    "220 nm SOI Strip Waveguide — Group Velocity Dispersion\n" + _subtitle,
    fontsize=12, fontweight="bold",
)

# ── Left: β₂ TE00 and TM00 ───────────────────────────────────────────────────
ax2a.plot(sweep_wl_nm, beta2_TE_ps2km,
          color=C_TE, lw=2.2, label=r"$\beta_2$ (TE00)")
ax2a.plot(sweep_wl_nm, beta2_TM_ps2km,
          color=C_TM, lw=2.2, ls="--", label=r"$\beta_2$ (TM00)")
ax2a.axhline(0, color=C_ZERO, lw=1.0, ls=":", alpha=0.8, label="β₂ = 0  (ZDW)")
ax2a.axvline(1550, color="gray", lw=0.8, ls=":", alpha=0.6, label="λ = 1550 nm")
if ZDW_TE:
    ax2a.axvline(ZDW_TE, color=C_ZDW, lw=1.2, ls="--",
                 label=f"ZDW ≈ {ZDW_TE:.0f} nm")
ax2a.fill_between(sweep_wl_nm, beta2_TE_ps2km, 0,
                   where=(beta2_TE_ps2km < 0), alpha=0.08, color=C_TE)
ax2a.fill_between(sweep_wl_nm, beta2_TE_ps2km, 0,
                   where=(beta2_TE_ps2km >= 0), alpha=0.08, color=C_TM)
ax2a.set_xlabel("Wavelength  λ  (nm)")
ax2a.set_ylabel(r"$\beta_2$  (ps²/km)")
ax2a.set_title(r"GVD Coefficient  $\beta_2$  — TE00 and TM00")
ax2a.legend(ncol=2, fontsize=8)
ax2a.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax2a.yaxis.set_minor_locator(ticker.AutoMinorLocator())

# ── Right: D TE00 and TM00 ────────────────────────────────────────────────────
ax2b.plot(sweep_wl_nm, D_TE_ps_nm_km,
          color=C_TE, lw=2.2, label=r"$D$ (TE00)")
ax2b.plot(sweep_wl_nm, D_TM_ps_nm_km,
          color=C_TM, lw=2.2, ls="--", label=r"$D$ (TM00)")
ax2b.axhline(0, color=C_ZERO, lw=1.0, ls=":", alpha=0.8, label="D = 0  (ZDW)")
ax2b.axvline(1550, color="gray", lw=0.8, ls=":", alpha=0.6, label="λ = 1550 nm")
if ZDW_TE:
    ax2b.axvline(ZDW_TE, color=C_ZDW, lw=1.2, ls="--",
                 label=f"ZDW ≈ {ZDW_TE:.0f} nm")

ax2b.set_xlabel("Wavelength  λ  (nm)")
ax2b.set_ylabel(r"Dispersion parameter  $D$  (ps/(nm·km))")
ax2b.set_title(r"Dispersion $D = -(2\pi c/\lambda^2)\,\beta_2$  — TE00 and TM00")
ax2b.legend(ncol=2, fontsize=8)
ax2b.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax2b.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig2.tight_layout()
_save(fig2, "HW4_SOI_GVD_vs_wavelength")
plt.show()


# ── Figure 3: Birefringence ────────────────────────────────────────────────────
fig3, ax3 = plt.subplots(figsize=(8, 5))
ax3.plot(sweep_wl_nm, birefringence, color="#009E73", lw=2.4,
         label=r"$B = n_\mathrm{eff,TE} - n_\mathrm{eff,TM}$")
ax3.axvline(1550, color="gray", lw=0.9, ls=":", alpha=0.8, label="λ = 1550 nm")
ax3.fill_between(sweep_wl_nm, birefringence, 0, alpha=0.12, color="#009E73")
ax3.set_xlabel("Wavelength  λ  (nm)")
ax3.set_ylabel(r"Modal birefringence  $B$")
ax3.set_ylim([0.5, 0.7])
ax3.set_title(
    f"Birefringence — 220 nm SOI Strip Waveguide\n"
    f"Si {WG_WIDTH_NM:.0f}×{WG_HEIGHT_NM:.0f} nm  │  "
    f"w/h ≈ {WG_WIDTH_NM/WG_HEIGHT_NM:.2f}  (asymmetric → high B)")
ax3.legend()
ax3.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax3.yaxis.set_minor_locator(ticker.AutoMinorLocator())
_biref_range = birefringence.max() - birefringence.min()
ax3.annotate(
    f"B(1550 nm) = {birefringence[_idx1550]:.4f}",
    xy=(1550, birefringence[_idx1550]),
    xytext=(1560, birefringence[_idx1550] + 3 * _biref_range),
    fontsize=9, color="#009E73",
    arrowprops=dict(arrowstyle="->", color="#009E73", lw=1.0),
)

fig3.tight_layout()
_save(fig3, "HW4_SOI_birefringence_vs_wavelength")
plt.show()


# ── Figure 4: 2×2 overview ────────────────────────────────────────────────────
fig4, axes4 = plt.subplots(2, 2, figsize=(13, 10))
fig4.suptitle(
    "HW4 Part 1 — 220 nm SOI Mode Analysis  (fixed index)\n" + _subtitle,
    fontsize=13, fontweight="bold",
)

ax = axes4[0, 0]
ax.plot(sweep_wl_nm, neff_TE, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, neff_TM, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$n_\mathrm{eff}$")
ax.set_title("Effective Index"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

ax = axes4[0, 1]
ax.plot(sweep_wl_nm, ng_TE, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, ng_TM, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$n_g$")
ax.set_title("Group Index"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

ax = axes4[1, 0]
ax.plot(sweep_wl_nm, beta2_TE_ps2km, color=C_TE, lw=2.0, label="TE00")
ax.plot(sweep_wl_nm, beta2_TM_ps2km, color=C_TM, lw=2.0, ls="--", label="TM00")
ax.axhline(0, color=C_ZERO, lw=0.9, ls=":", alpha=0.8)
if ZDW_TE:
    ax.axvline(ZDW_TE, color=C_ZDW, lw=1.2, ls="--",
               label=f"ZDW ≈ {ZDW_TE:.0f} nm")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"$\beta_2$ (ps²/km)")
ax.set_title(r"GVD $\beta_2$"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

ax = axes4[1, 1]
ax.plot(sweep_wl_nm, birefringence, color="#009E73", lw=2.0,
        label=r"$B = n_\mathrm{eff,TE} - n_\mathrm{eff,TM}$")
ax.set_xlabel("λ (nm)"); ax.set_ylabel(r"Birefringence $B$")
ax.set_title("Modal Birefringence"); ax.legend()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

fig4.tight_layout()
_save(fig4, "HW4_SOI_overview_4panel")
plt.show()

print(f"\n  All figures saved to:  {DATA_DIR}")
print(f"  HDF5 data file:        {HDF5_PATH}")
```

```python
"""
HW4 Part 2 -- Thermo-Optic device layouts WITH EDGE-COUPLER I/O (generic PDK)
============================================================================

Same two thermo-optically tuned devices as before (Part 2 of the report),
but every device is now book-ended by an *edge-coupler* transition on both
the input and the output facet:

        multimode straight  ->  adiabatic taper  ->  single-mode straight
        (wide facet WG)         (MM width -> SM)      (feeds the device)

This is the standard edge / butt-coupling front end for a 220 nm SOI chip:
a wide (multimode) waveguide at the cleaved facet relaxes the alignment
tolerance to a lensed fibre, and the linear taper adiabatically squeezes the
spot down to the single-mode routing width before it reaches the active
device.  The same structure, mirrored, is placed at the output.

Devices:
    * Thermo-optic Mach-Zehnder modulator (MZM), FSR = 10 nm
        - unbalanced MZI, arm imbalance  dL = lambda^2 / (ng * FSR)
        - metal micro-heater on the top arm
    * Thermo-optic micro-ring modulator (MRM), FSR = 10 nm
        - all-pass single-bus ring, R = lambda^2 / (2*pi*ng*FSR)
        - resistive heater wrapped over the ring

Outputs (in ./gds_out/):
    HW4_TO_MZM.gds, HW4_TO_MRM.gds, HW4_Part2_combined.gds  (+ matching PNGs)

The individual device GDS names are kept identical to the previous version so
the rest of the workflow (and the Part-4 integration cell) keeps working; the
only change is that each device now contains its edge-coupler front ends.
"""

from __future__ import annotations
import math
import pathlib
import gdsfactory as gf
from gdsfactory.gpdk import get_generic_pdk

# Activate the GDSfactory generic PDK (WG = (1,0), heater + metal routing).
get_generic_pdk().activate()

# -----------------------------------------------------------------------------
# 1. Physics-driven design parameters (from Part 1 / Part 2 of the report)
# -----------------------------------------------------------------------------
LAMBDA0 = 1.550          # design wavelength [um]
NG_TE   = 4.3            # TE group index at 1550 nm (Part 1 FDE result)
FSR     = 0.010          # target free spectral range [um]  (= 10 nm)

# Arm imbalance for the MZM:  FSR = lambda^2 / (ng * dL)  ->  dL = lambda^2/(ng*FSR)
DELTA_L_MZM = LAMBDA0**2 / (NG_TE * FSR)            # ~= 55.9 um

# Ring radius for the MRM:  L_rt = lambda^2/(ng*FSR) = 2*pi*R
L_RT        = LAMBDA0**2 / (NG_TE * FSR)            # round-trip length ~= 55.9 um
RADIUS_MRM  = L_RT / (2.0 * math.pi)               # ~= 8.9 um

print(f"[design] dL (MZM)     = {DELTA_L_MZM:7.2f} um")
print(f"[design] L_rt (ring)  = {L_RT:7.2f} um")
print(f"[design] R (MRM)      = {RADIUS_MRM:7.2f} um")

# Heater geometry (used only to size the active straight sections)
HEATER_LEN = 100.0       # heater length on the MZM top arm [um]
RING_GAP   = 0.20        # bus-ring coupling gap [um] (>= generic DRC min)

# -----------------------------------------------------------------------------
# 1b. Edge-coupler (butt-coupling) front-end geometry
# -----------------------------------------------------------------------------
WG_LAYER     = (1, 0)    # generic-PDK strip waveguide layer
SM_WIDTH     = 0.50      # single-mode routing width [um] (220 nm SOI strip)
MM_WIDTH     = 3.00      # multimode facet width [um] (relaxed fibre alignment)
MM_LENGTH    = 15.0      # multimode straight at the facet [um]
TAPER_LENGTH = 40.0      # adiabatic MM -> SM linear taper length [um]
SM_LEAD      = 5.0       # single-mode lead straight into the device [um]


# -----------------------------------------------------------------------------
# 2. Edge-coupler builder + generic "wrap any device with edge couplers" helper
# -----------------------------------------------------------------------------
@gf.cell
def edge_coupler_io(
    mm_width: float = MM_WIDTH,
    sm_width: float = SM_WIDTH,
    mm_length: float = MM_LENGTH,
    taper_length: float = TAPER_LENGTH,
    sm_length: float = SM_LEAD,
) -> gf.Component:
    """Facet edge coupler: MM straight -> MM->SM taper -> SM lead straight.

    Ports
    -----
    o1 : multimode facet port  (wide, sits at the chip edge)
    o2 : single-mode port       (mates with the device input/output)
    """
    c = gf.Component()
    mm = c << gf.components.straight(length=mm_length, width=mm_width, cross_section="strip")
    tp = c << gf.components.taper(length=taper_length, width1=mm_width,
                                  width2=sm_width, layer=WG_LAYER)
    sm = c << gf.components.straight(length=sm_length, width=sm_width, cross_section="strip")
    tp.connect("o1", mm.ports["o2"])
    sm.connect("o1", tp.ports["o2"])
    c.add_port("o1", port=mm.ports["o1"])   # multimode facet
    c.add_port("o2", port=sm.ports["o2"])   # single-mode device side
    return c


@gf.cell
def with_edge_couplers(device: gf.Component) -> gf.Component:
    """Book-end any 2-port (o1/o2) optical device with edge couplers on both facets."""
    c = gf.Component()
    dev = c << device
    ec_in = c << edge_coupler_io()
    ec_out = c << edge_coupler_io()

    # SM side (o2) of each coupler mates with the device optical ports.
    ec_in.connect("o2", dev.ports["o1"])
    ec_out.connect("o2", dev.ports["o2"])

    # Expose the multimode facet ports at the top level.
    c.add_port("o1", port=ec_in.ports["o1"])
    c.add_port("o2", port=ec_out.ports["o1"])

    # Keep every electrical port (heater / pads) available at the top level.
    for p in dev.ports:
        if p.port_type == "electrical":
            c.add_port(p.name, port=p)
    return c


# -----------------------------------------------------------------------------
# 3. Device builders (now edge-coupled)
# -----------------------------------------------------------------------------
@gf.cell
def to_mzm_core(
    delta_length: float = DELTA_L_MZM,
    heater_length: float = HEATER_LEN,
) -> gf.Component:
    """Unbalanced MZI with a metal heater on the top arm (thermo-optic MZM core)."""
    return gf.components.mzi_phase_shifter_top_heater_metal(
        delta_length=delta_length,   # arm imbalance -> sets the 10 nm FSR
        length_x=heater_length,      # heated straight section length
        length_y=2.0,
        straight_x_top="straight_heater_metal",
    )


@gf.cell
def to_mrm_core(
    radius: float = RADIUS_MRM,
    gap: float = RING_GAP,
) -> gf.Component:
    """All-pass single-bus ring with an over-the-ring metal heater (MRM core)."""
    return gf.components.ring_single_heater(
        radius=radius,               # sets the 10 nm FSR (round-trip = 2*pi*R)
        gap=gap,
        length_x=0.5,                # short coupling straight -> near-circular
        length_y=0.5,
    )


@gf.cell
def to_mzm() -> gf.Component:
    """Thermo-optic MZM with edge-coupler I/O on both facets."""
    return with_edge_couplers(to_mzm_core())


@gf.cell
def to_mrm() -> gf.Component:
    """Thermo-optic micro-ring modulator with edge-coupler I/O on both facets."""
    return with_edge_couplers(to_mrm_core())


@gf.cell
def hw4_part2_combined(pitch_y: float = 140.0) -> gf.Component:
    """Both edge-coupled devices in one top cell for the combined .gds deliverable."""
    top = gf.Component()
    mzm = top << to_mzm()
    mrm = top << to_mrm()

    mzm.movey(0)
    mrm.movey(-pitch_y)

    top.add_port("mzm_o1", port=mzm.ports["o1"])
    top.add_port("mzm_o2", port=mzm.ports["o2"])
    top.add_port("mrm_o1", port=mrm.ports["o1"])
    top.add_port("mrm_o2", port=mrm.ports["o2"])
    return top


# -----------------------------------------------------------------------------
# 4. Build, write GDS, render PNGs, and open each layout in KLayout
# -----------------------------------------------------------------------------
def main() -> None:
    out = pathlib.Path("gds_out")
    out.mkdir(exist_ok=True)

    devices = {
        "HW4_TO_MZM":         to_mzm(),
        "HW4_TO_MRM":         to_mrm(),
        "HW4_Part2_combined": hw4_part2_combined(),
    }

    for name, comp in devices.items():
        gds_path = out / f"{name}.gds"
        comp.write_gds(gds_path)
        print(f"[gds ] wrote {gds_path}")

        try:
            gf.show(comp)
            print(f"[klayout] opened {name}")
        except Exception as exc:
            print(f"[klayout] skipped {name}: {exc}")

        try:
            import matplotlib
            matplotlib.use("Agg")
            import matplotlib.pyplot as plt
            ax = comp.plot()
            fig = ax.get_figure() if hasattr(ax, "get_figure") else plt.gcf()
            fig.savefig(out / f"{name}.png", dpi=200, bbox_inches="tight")
            plt.close(fig)
            print(f"[png ] wrote {out / (name + '.png')}")
        except Exception as exc:
            print(f"[png ] skipped {name}: {exc}")


if __name__ == "__main__":
    main()
```

```python
"""
HW4 Part 3 -- Electro-Optic device layouts (GDSfactory generic PDK)
===================================================================

Companion cell to the Part 2 script. Same generic PDK, same ./gds_out/ folder,
disjoint names (eo_*, hw4_part3_*, main_part3) so both cells coexist.

    * EO Mach-Zehnder modulator: unbalanced MZI, both arms = 'straight_pn'
      PN-junction depletion phase shifters (Soref-Bennett free-carrier effect).
    * EO micro-ring modulator (FSR = 10 nm): all-pass ring with a lateral PN
      junction; reuses the Part 1 group index for the radius.

Outputs (in ./gds_out/):
    HW4_EO_MZM.gds, HW4_EO_MRM.gds, HW4_Part3_combined.gds
    HW4_EO_MZM.png, HW4_EO_MRM.png, HW4_Part3_combined.png
"""

from __future__ import annotations
import math
import pathlib
import gdsfactory as gf
from gdsfactory.gpdk import get_generic_pdk

get_generic_pdk().activate()

# -----------------------------------------------------------------------------
# 1. Physics-driven design parameters (consistent with Part 1 / Part 2)
# -----------------------------------------------------------------------------
LAMBDA0 = 1.550          # design wavelength [um]
NG_TE   = 4.3            # TE group index at 1550 nm (match Part 2; use 4.16 for 9.2 um)
FSR     = 0.010          # target free spectral range [um]  (= 10 nm)

DELTA_L_EO = LAMBDA0**2 / (NG_TE * FSR)            # MZM arm imbalance  ~= 55.9 um
L_RT_EO    = LAMBDA0**2 / (NG_TE * FSR)            # ring round-trip    ~= 55.9 um
RADIUS_EO  = L_RT_EO / (2.0 * math.pi)             # ring radius        ~= 8.9 um

PN_LENGTH = 300.0        # drawn PN phase-shifter length [um] (real design ~3 mm)
RING_GAP  = 0.20         # bus-ring coupling gap [um] (>= generic DRC min)

print(f"[design] dL (EO-MZM)  = {DELTA_L_EO:7.2f} um")
print(f"[design] L_rt (ring)  = {L_RT_EO:7.2f} um")
print(f"[design] R (EO-MRM)   = {RADIUS_EO:7.2f} um")


# -----------------------------------------------------------------------------
# 2. Component builders (generic PDK)
# -----------------------------------------------------------------------------
@gf.cell
def eo_mzm(
    delta_length: float = DELTA_L_EO,
    pn_length: float = PN_LENGTH,
) -> gf.Component:
    """Unbalanced MZI with PN-junction phase shifters on both arms (EO-MZM)."""
    return gf.components.mzi(
        delta_length=delta_length,    # sets the 10 nm FSR (same as Part 2)
        length_x=pn_length,           # PN phase-shifter length
        length_y=2.0,
        straight_x_top="straight_pn",
        straight_x_bot="straight_pn",  # <-- was 'straight_x_bottom' (fixed)
    )


@gf.cell
def eo_mrm(
    radius: float = RADIUS_EO,
    gap: float = RING_GAP,
) -> gf.Component:
    """All-pass single-bus ring with a lateral PN junction along the ring (EO-MRM)."""
    # Try the dedicated doped ring first; fall back across version differences.
    try:
        return gf.components.ring_single_pn(radius=radius, gap=gap)
    except (AttributeError, TypeError):
        pass
    try:
        return gf.components.ring_single(
            radius=radius, gap=gap, length_x=0.5, length_y=0.5,
            cross_section="pn",
        )
    except (TypeError, ValueError, KeyError):
        # Last resort: plain strip ring (doping added later in PhotonForge).
        return gf.components.ring_single(
            radius=radius, gap=gap, length_x=0.5, length_y=0.5,
        )


@gf.cell
def hw4_part3_combined(pitch_y: float = 140.0) -> gf.Component:
    """Both EO devices in one top cell for the Part 3 .gds deliverable."""
    top = gf.Component()
    mzm = top << eo_mzm()
    mrm = top << eo_mrm()

    mzm.movey(0)
    mrm.movey(-pitch_y)

    top.add_port("eo_mzm_o1", port=mzm.ports["o1"])
    top.add_port("eo_mzm_o2", port=mzm.ports["o2"])
    top.add_port("eo_mrm_o1", port=mrm.ports["o1"])
    top.add_port("eo_mrm_o2", port=mrm.ports["o2"])
    return top


# -----------------------------------------------------------------------------
# 3. Build, write GDS, render PNGs, open each layout in KLayout (same as Part 2)
# -----------------------------------------------------------------------------
def main_part3() -> None:
    out = pathlib.Path("gds_out")           # SAME storage folder as Part 2
    out.mkdir(exist_ok=True)

    devices = {
        "HW4_EO_MZM":          eo_mzm(),
        "HW4_EO_MRM":          eo_mrm(),
        "HW4_Part3_combined":  hw4_part3_combined(),
    }

    for name, comp in devices.items():
        gds_path = out / f"{name}.gds"
        comp.write_gds(gds_path)
        print(f"[gds ] wrote {gds_path}")

        try:
            gf.show(comp)
            print(f"[klayout] opened {name}")
        except Exception as exc:
            print(f"[klayout] skipped {name}: {exc}")

        try:
            import matplotlib
            matplotlib.use("Agg")
            import matplotlib.pyplot as plt
            ax = comp.plot()
            fig = ax.get_figure() if hasattr(ax, "get_figure") else plt.gcf()
            fig.savefig(out / f"{name}.png", dpi=200, bbox_inches="tight")
            plt.close(fig)
            print(f"[png ] wrote {out / (name + '.png')}")
        except Exception as exc:
            print(f"[png ] skipped {name}: {exc}")


if __name__ == "__main__":
    main_part3()
```
